### Generating a list in python containing monthly dates from 2011-01-01 to 2021-12-31 that it will be used as an index

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

start_date = datetime(2011, 1, 1)  # Initial date
end_date = datetime(2021, 12, 31)  # Final date

dates = []                         # Empty list that it will be used to store the monthly date created 
current_date = start_date

while current_date <= end_date:
    # Inside the loop we create a current_date variable that starts with start_date and increment this variable with the timedelta function
    dates.append(current_date.strftime('%Y-%m-%d'))
    current_date += relativedelta(months=1)

print(dates)
```

```python
# Creating a dataframe using the data from cdi and the generated dates from algorithm as index
my_cdi = pd.DataFrame(teste,index = dates, columns=['CDI'])
```
